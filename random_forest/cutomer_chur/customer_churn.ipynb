```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
X=df.drop(columns=['churn'])
y=df['churn']
```

```python
X.shape,y.shape
```

```python
X.columns
```

```python
y.name
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
oe=OrdinalEncoder()
X_train[['country','gender']]=oe.fit_transform(X_train[['country','gender']])
X_test[['country','gender']]=oe.fit_transform(X_test[['country','gender']])
```

```python
X_train.head()
```

```python
X_train.head()
```

```python
nor=Normalizer()
X_train=nor.fit_transform(X_train)
X_test=nor.transform(X_test)
```

```python
model=RandomForestClassifier(criterion='entropy')
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
print(f"{model.score(X_test,y_test)*100 : .2f} %")
```
